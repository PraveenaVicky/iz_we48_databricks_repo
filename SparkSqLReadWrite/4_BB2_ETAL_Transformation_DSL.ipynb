```python
cust_raw_df=spark.read.csv("/Volumes/izwe48catalog/we48db/staging/BB2/custsmodified",header=False,sep=",").toDF("custid","fname","lname","age","prof")

cust_raw_df.show(2)
```

```python
# view the schema 
print("Schema of cust ")
cust_raw_df.printSchema()

# all coulumn names in the dataframe
print("Column names of cust ")
col_names=cust_raw_df.columns  # list[str]

print(col_names)


# get col name and datatype 

col_name_type=cust_raw_df.dtypes   # list[(col_name,dtype)]
# [('custid', 'string'), ('fname', 'string'), ('lname', 'string'), ('age', 'string'), ('prof', 'string')]

print(col_name_type)

# need columns which is folliwing string
for i in col_name_type:
  if i[1]=="string":
      print(i[0])

# get the schema 

cust_schema=cust_raw_df.schema


print(cust_schema)

```

```python
# important 
# whther given DF has duplicates 

# get the total count of the DF


print(cust_raw_df.count())   # action 

# row level / record level  unique count 

print(cust_raw_df.distinct().count())  # de duplication on record level 


# record level de duplication using DropDuplicates
print(cust_raw_df.dropDuplicates().count())


cust_raw_df.filter("custid='4000001'").show(5)
cust_raw_df.dropDuplicates().filter("custid='4000001'").show(5)
cust_raw_df.distinct().filter("custid='4000001'").show(5)




```
